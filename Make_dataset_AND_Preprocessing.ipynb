## Resize ảnh về (640,640)
##### <i> - Kích thước(640,640) phù hợp để train yolov8</i>
##### <i> - Sử dụng đánh nhãn bán tự động</i>

```python
import os
from PIL import Image

def resize_images(input_folder, output_folder, target_size=(640, 640)):
    # Tạo thư mục đầu ra nếu nó chưa tồn tại
    os.makedirs(output_folder, exist_ok=True)

    # Lặp qua từng file trong thư mục input_folder
    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)

        # Kiểm tra nếu là file ảnh
        if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp')):
            try:
                # Đọc ảnh từ đường dẫn
                image = Image.open(input_path)

                # Resize ảnh về kích thước mong muốn
                resized_image = image.resize(target_size)

                # Lưu ảnh đã resize vào thư mục đầu ra
                output_path = os.path.join(output_folder, filename)
                resized_image.save(output_path)
                
                print(f"Đã resize và lưu '{filename}' thành công.")
            except Exception as e:
                print(f"Lỗi khi xử lý file '{filename}': {e}")

# Đường dẫn đến thư mục chứa các ảnh cần resize
input_folder = 'test'

# Đường dẫn đến thư mục để lưu các ảnh đã resize
output_folder = 'test'

# Kích thước mới cho các ảnh sau khi resize
target_size = (640, 640)

# Gọi hàm để resize các ảnh trong thư mục input_folder và lưu vào output_folder
resize_images(input_folder, output_folder, target_size)
```

```python
from pathlib import Path
from tqdm import tqdm
from ultralytics import YOLO

class_name = ['microsleep','yawn']
def predict_images_in_folder(input_folder, weights='last.pt'):
    # Khởi tạo mô hình YOLOv5
    model = YOLO(weights)

    # Lấy danh sách các đường dẫn tới các file ảnh trong thư mục input_folder
    image_paths = list(Path(input_folder).glob('*.jpg'))  # Lấy tất cả các file jpg trong thư mục

    # Lặp qua từng file ảnh để dự đoán và lưu kết quả
    for image_path in tqdm(image_paths, desc='Predicting'):
        # Dự đoán trên ảnh và không lưu file text (.txt)
        predictions = model(str(image_path), save_txt=None)

        # Tạo đường dẫn tới file text kết quả (cùng tên với file ảnh nhưng có đuôi .txt)
        txt_filename = image_path.with_suffix('.txt')

        # Ghi dữ liệu dự đoán vào file text
        with open(txt_filename, 'w') as file:
            for idx, prediction in enumerate(predictions[0].boxes.xywhn):
                cls = int(predictions[0].boxes.cls[idx].item())
                # x, y, w, h = prediction[:4].tolist()  # Lấy các giá trị x, y, w, h từ prediction
                # file.write(f"{cls} {x} {y} {w} {h}\n")
                file.write(f"{class_name[cls]} \n")

if __name__ == "__main__":
    # Đường dẫn đến thư mục chứa các ảnh bạn muốn dự đoán
    input_folder = r"D:\University\DATN\Drowsiness Detection\Data\DataDectection"

    # File trọng số của mô hình YOLOv5 (ví dụ: 'last.pt' hoặc đường dẫn tới file trọng số)
    weights_path = "last.pt"

    # Thực hiện dự đoán trên tất cả các ảnh trong thư mục và lưu kết quả vào các file text
    predict_images_in_folder(input_folder, weights_path)
```

## Xóa file rỗng

```python
import os

def is_empty_file(file_path):
    """Kiểm tra xem file có rỗng hay không."""
    return os.path.exists(file_path) and os.stat(file_path).st_size == 0

def remove_empty_txt_files(folder_path):
    """Xóa các file .txt rỗng trong thư mục."""
    # Lấy danh sách các file .txt trong thư mục
    txt_files = [f for f in os.listdir(folder_path) if f.endswith('.txt')]
    
    # Duyệt qua từng file và kiểm tra xóa nếu file rỗng
    for txt_file in txt_files:
        file_path = os.path.join(folder_path, txt_file)
        
        if is_empty_file(file_path):
            img,extension = os.path.splitext(txt_file)
            image = img+".jpg"
            # print(os.path.splitext(txt_file))
            img_path = os.path.join(folder_path, image)
            os.remove(file_path)
            os.remove(img_path)
            print(f"Đã xóa file rỗng: {txt_file} and {image}")

# Đường dẫn đến thư mục chứa các file .txt
folder_path = r'D:\University\DATN\Drowsiness Detection\Data\DataDectection'

# Xóa các file .txt rỗng trong thư mục
remove_empty_txt_files(folder_path)
```

## Xóa lặp lại

```python
import os

def remove_duplicate_lines_in_files(folder_path):
    """Xử lý tất cả các file .txt trong thư mục để xóa các dòng lặp lại."""
    # Lặp qua tất cả các file trong thư mục
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.txt'):  # Chỉ xử lý các file có đuôi .txt
            file_path = os.path.join(folder_path, file_name)
            process_single_txt_file(file_path)

def process_single_txt_file(file_path):
    """Xóa các dòng lặp lại trong một file .txt."""
    lines_seen = set()

    # Đọc từng dòng trong file và xử lý
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    with open(file_path, 'w', encoding='utf-8') as file:
        for line in lines:
            line = line.strip()  # Loại bỏ khoảng trắng và dấu xuống dòng
            if line not in lines_seen:
                file.write(line + '\n')
                lines_seen.add(line)

    print(f"Đã xử lý file '{file_path}' để xóa các dòng lặp lại.")

# Đường dẫn đến thư mục chứa các file .txt
folder_path = r'D:\University\DATN\Drowsiness Detection\Data\DataDectection'

# Xử lý tất cả các file .txt trong thư mục để xóa các dòng lặp lại
remove_duplicate_lines_in_files(folder_path)
```

```python
import os

def process_text_file(file_path):
    """Đọc và xử lý nội dung của một file văn bản."""
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()  # Đọc tất cả các dòng trong file

    # Nối các dòng lại thành một chuỗi duy nhất
    content = ''.join(lines)

    # Kiểm tra và thực hiện sửa đổi nội dung theo điều kiện
    if "microsleep" in content and "yawn" in content:
        # Thay thế thành "yawn microsleep"
        content = "yawn microsleep"

    elif "yawn" in content and "microsleep" in content:
        # Thay thế thành "yawn microsleep"
        content = "yawn microsleep"

    # Ghi lại nội dung đã được sửa đổi vào file
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(content)

def process_text_files_in_folder(folder_path):
    """Duyệt và xử lý các file văn bản trong thư mục được chỉ định."""
    # Lấy danh sách các file văn bản (.txt) trong thư mục
    txt_files = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

    # Duyệt qua từng file và thực hiện xử lý
    for txt_file in txt_files:
        file_path = os.path.join(folder_path, txt_file)
        process_text_file(file_path)
        print(f"Đã xử lý file: {txt_file}")

if __name__ == "__main__":
    # Thay đổi đường dẫn đến thư mục chứa các file văn bản (.txt)
    folder_path = r'D:\University\DATN\Drowsiness Detection\Data\DataDectection'

    # Duyệt và xử lý các file văn bản trong thư mục folder_path
    process_text_files_in_folder(folder_path)
```

## Head pose

```python
import cv2
import mediapipe as mp
import numpy as np

############## PARAMETERS #######################################################

# Set these values to show/hide certain vectors of the estimation
draw_gaze = True
draw_full_axis = True
draw_headpose = False

# Gaze Score multiplier (Higher multiplier = Gaze affects headpose estimation more)
x_score_multiplier = 10
y_score_multiplier = 10

# Threshold of how close scores should be to average between frames
threshold = .3

#################################################################################

mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=False,
    refine_landmarks=True,
    max_num_faces=2,
    min_detection_confidence=0.5)
cap = cv2.VideoCapture(0)

face_3d = np.array([
    [0.0, 0.0, 0.0],            # Nose tip
    [0.0, -330.0, -65.0],       # Chin
    [-225.0, 170.0, -135.0],    # Left eye left corner
    [225.0, 170.0, -135.0],     # Right eye right corner
    [-150.0, -150.0, -125.0],   # Left Mouth corner
    [150.0, -150.0, -125.0]     # Right mouth corner
    ], dtype=np.float64)

# Reposition left eye corner to be the origin
leye_3d = np.array(face_3d)
leye_3d[:,0] += 225
leye_3d[:,1] -= 175
leye_3d[:,2] += 135

# Reposition right eye corner to be the origin
reye_3d = np.array(face_3d)
reye_3d[:,0] -= 225
reye_3d[:,1] -= 175
reye_3d[:,2] += 135

# Gaze scores from the previous frame
last_lx, last_rx = 0, 0
last_ly, last_ry = 0, 0

while cap.isOpened():
    success, img = cap.read()

    # Flip + convert img from BGR to RGB
    img = cv2.cvtColor(cv2.flip(img, 1), cv2.COLOR_BGR2RGB)

    # To improve performance
    img.flags.writeable = False
    
    # Get the result
    results = face_mesh.process(img)
    img.flags.writeable = True
    
    # Convert the color space from RGB to BGR
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

    (img_h, img_w, img_c) = img.shape
    face_2d = []

    if not results.multi_face_landmarks:
      continue 

    for face_landmarks in results.multi_face_landmarks:
        face_2d = []
        for idx, lm in enumerate(face_landmarks.landmark):
            # Convert landmark x and y to pixel coordinates
            x, y = int(lm.x * img_w), int(lm.y * img_h)

            # Add the 2D coordinates to an array
            face_2d.append((x, y))
        
        # Get relevant landmarks for headpose estimation
        face_2d_head = np.array([
            face_2d[1],      # Nose
            face_2d[199],    # Chin
            face_2d[33],     # Left eye left corner
            face_2d[263],    # Right eye right corner
            face_2d[61],     # Left mouth corner
            face_2d[291]     # Right mouth corner
        ], dtype=np.float64)

        face_2d = np.asarray(face_2d)

        # Calculate left x gaze score
        if (face_2d[243,0] - face_2d[130,0]) != 0:
            lx_score = (face_2d[468,0] - face_2d[130,0]) / (face_2d[243,0] - face_2d[130,0])
            if abs(lx_score - last_lx) < threshold:
                lx_score = (lx_score + last_lx) / 2
            last_lx = lx_score

        # Calculate left y gaze score
        if (face_2d[23,1] - face_2d[27,1]) != 0:
            ly_score = (face_2d[468,1] - face_2d[27,1]) / (face_2d[23,1] - face_2d[27,1])
            if abs(ly_score - last_ly) < threshold:
                ly_score = (ly_score + last_ly) / 2
            last_ly = ly_score

        # Calculate right x gaze score
        if (face_2d[359,0] - face_2d[463,0]) != 0:
            rx_score = (face_2d[473,0] - face_2d[463,0]) / (face_2d[359,0] - face_2d[463,0])
            if abs(rx_score - last_rx) < threshold:
                rx_score = (rx_score + last_rx) / 2
            last_rx = rx_score

        # Calculate right y gaze score
        if (face_2d[253,1] - face_2d[257,1]) != 0:
            ry_score = (face_2d[473,1] - face_2d[257,1]) / (face_2d[253,1] - face_2d[257,1])
            if abs(ry_score - last_ry) < threshold:
                ry_score = (ry_score + last_ry) / 2
            last_ry = ry_score

        # The camera matrix
        focal_length = 1 * img_w
        cam_matrix = np.array([ [focal_length, 0, img_h / 2],
                                [0, focal_length, img_w / 2],
                                [0, 0, 1]])

        # Distortion coefficients 
        dist_coeffs = np.zeros((4, 1), dtype=np.float64)

        # Solve PnP
        _, l_rvec, l_tvec = cv2.solvePnP(leye_3d, face_2d_head, cam_matrix, dist_coeffs, flags=cv2.SOLVEPNP_ITERATIVE)
        _, r_rvec, r_tvec = cv2.solvePnP(reye_3d, face_2d_head, cam_matrix, dist_coeffs, flags=cv2.SOLVEPNP_ITERATIVE)


        # Get rotational matrix from rotational vector
        l_rmat, _ = cv2.Rodrigues(l_rvec)
        r_rmat, _ = cv2.Rodrigues(r_rvec)


        # [0] changes pitch
        # [1] changes roll
        # [2] changes yaw
        # +1 changes ~45 degrees (pitch down, roll tilts left (counterclockwise), yaw spins left (counterclockwise))

        # Adjust headpose vector with gaze score
        l_gaze_rvec = np.array(l_rvec)
        l_gaze_rvec[2][0] -= (lx_score-.5) * x_score_multiplier
        l_gaze_rvec[0][0] += (ly_score-.5) * y_score_multiplier

        r_gaze_rvec = np.array(r_rvec)
        r_gaze_rvec[2][0] -= (rx_score-.5) * x_score_multiplier
        r_gaze_rvec[0][0] += (ry_score-.5) * y_score_multiplier

        # --- Projection ---

        # Get left eye corner as integer
        l_corner = face_2d_head[2].astype(np.int32)

        # Project axis of rotation for left eye
        axis = np.float32([[-100, 0, 0], [0, 100, 0], [0, 0, 300]]).reshape(-1, 3)
        l_axis, _ = cv2.projectPoints(axis, l_rvec, l_tvec, cam_matrix, dist_coeffs)
        l_gaze_axis, _ = cv2.projectPoints(axis, l_gaze_rvec, l_tvec, cam_matrix, dist_coeffs)

        # Draw axis of rotation for left eye
        if draw_headpose:
            if draw_full_axis:
                cv2.line(img, l_corner, tuple(np.ravel(l_axis[0]).astype(np.int32)), (200,200,0), 3)
                cv2.line(img, l_corner, tuple(np.ravel(l_axis[1]).astype(np.int32)), (0,200,0), 3)
            cv2.line(img, l_corner, tuple(np.ravel(l_axis[2]).astype(np.int32)), (0,200,200), 3)

        if draw_gaze:
            if draw_full_axis:
                cv2.line(img, l_corner, tuple(np.ravel(l_gaze_axis[0]).astype(np.int32)), (255,0,0), 3)
                cv2.line(img, l_corner, tuple(np.ravel(l_gaze_axis[1]).astype(np.int32)), (0,255,0), 3)
            cv2.line(img, l_corner, tuple(np.ravel(l_gaze_axis[2]).astype(np.int32)), (0,0,255), 3)

        
    
        # Get left eye corner as integer
        r_corner = face_2d_head[3].astype(np.int32)

        # Get left eye corner as integer
        r_axis, _ = cv2.projectPoints(axis, r_rvec, r_tvec, cam_matrix, dist_coeffs)
        r_gaze_axis, _ = cv2.projectPoints(axis, r_gaze_rvec, r_tvec, cam_matrix, dist_coeffs)

        # Draw axis of rotation for left eye
        if draw_headpose:
            if draw_full_axis:
                cv2.line(img, r_corner, tuple(np.ravel(r_axis[0]).astype(np.int32)), (200,200,0), 3)
                cv2.line(img, r_corner, tuple(np.ravel(r_axis[1]).astype(np.int32)), (0,200,0), 3)
            cv2.line(img, r_corner, tuple(np.ravel(r_axis[2]).astype(np.int32)), (0,200,200), 3)

        if draw_gaze:
            if draw_full_axis:
                cv2.line(img, r_corner, tuple(np.ravel(r_gaze_axis[0]).astype(np.int32)), (255,0,0), 3)
                cv2.line(img, r_corner, tuple(np.ravel(r_gaze_axis[1]).astype(np.int32)), (0,255,0), 3)
            cv2.line(img, r_corner, tuple(np.ravel(r_gaze_axis[2]).astype(np.int32)), (0,0,255), 3)
                


    cv2.imshow('Head Pose Estimation', img)

    if cv2.waitKey(5) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
